# Read HTML

Target Website: https://en.wikipedia.org/wiki/List_of_The_Simpsons_episodes_(seasons_1%E2%80%9320)

```python
simpsons_url = 'https://en.wikipedia.org/wiki/List_of_The_Simpsons_episodes_(seasons_1%E2%80%9320)'
simpsons = pd.read_html(simpsons_url)
print(len(simpsons))
```

```python
simpsons[1].head()
```

```python
def read_html_table(url, table_index=0, columns=0):
    # Fetch the HTML content
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'lxml')    
    # Find all tables in the HTML
    tables = soup.find_all('table')
    if not tables:
        raise ValueError("No tables found in the HTML")
    if table_index >= len(tables):
        raise IndexError(f"Table index {table_index} is out of range. Only {len(tables)} tables found.")
    # Select the desired table
    table = tables[table_index]
    # Extract headers
    headers = [th.text.strip() for th in table.find_all('th')]
    headers = headers[:columns]
    # Extract rows
    rows = []
    for tr in table.find_all('tr'):
        row = [td.text.strip() for td in tr.find_all('td')]
        if row:  # Skip empty rows
            rows.append(row)
    # Create Polars DataFrame
    df = pl.DataFrame(rows, schema=headers if headers else None)
    return df

# Usage
df = read_html_table(url=simpsons_url, table_index=1, columns=7)
```
